```python
import nltk
from nltk.collocations import ngrams
from nltk.corpus import udhr
from nltk.stem import PorterStemmer 
import re
from nltk.stem import WordNetLemmatizer
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
from french_lefff_lemmatizer.french_lefff_lemmatizer import FrenchLefffLemmatizer


english = udhr.raw('English-Latin1')
french = udhr.raw('French_Francais-Latin1')
italian = udhr.raw('Italian_Italiano-Latin1')
spanish = udhr.raw('Spanish_Espanol-Latin1')
```

```python
english_train, english_dev = english[0:1000], english[1000:1100]
french_train, french_dev = french[0:1000], french[1000:1100]
italian_train, italian_dev = italian[0:1000], italian[1000:1100]
spanish_train, spanish_dev = spanish[0:1000], spanish[1000:1100]
```

```python
english_test = udhr.words('English-Latin1')[0:1000]
french_test  = udhr.words('French_Francais-Latin1')[0:1000]
italian_test = udhr.words('Italian_Italiano-Latin1')[0:1000]
spanish_test = udhr.words('Spanish_Espanol-Latin1')[0:1000]
```

```python
#total number of words depend on only the training data,currently taking english as example

total_words = len(english_train)
def preprocessing(data,data_type,language):
    if data_type == "train":
        data = data.lower()
    stemmed_words = []
    lematize_words = []
       
    if language == "english":
        if data_type == "train":
            wordList   = re.sub('[^a-z\ \']+', " ",  data).split()
        else:
            wordList = [w.lower() for w in data if re.sub('[^a-z\ \']+', " ",  w) not in " " ]
        lemmatizer = WordNetLemmatizer()
    elif language == "french":
        print("french")
        lemmatizer = FrenchLemmatizer()
    elif language == "italian":
        print("italian")
#         lemmatizer = 
        for word in wordList:
            lematize_words.append(lemmatizer.lemmatize(word))
    for word in wordList:
        lematize_words.append(lemmatizer.lemmatize(word))
    Stemmer = SnowballStemmer(language)
    for w in lematize_words:
        stemmed_words.append(Stemmer.stem(word))
    stop_words = set(stopwords.words(language))
    stop_words = [w for w in lematize_words if not w in stop_words]
    return stop_words

english_train = preprocessing(english_train,"train","english")
english_dev = preprocessing(english_dev,"train","english")
english_test = preprocessing(english_test,"test","english")
#     stop_words_english = [w for w in wordList if not w in stop_words_english] 
#     stop_words_english = set(stopwords.words('english'))
#     stop_words_french  = set(stopwords.words('french'))
#     stop_words_italian = set(stopwords.words('italian'))
#     stop_words_spanish = set(stopwords.words('spanish'))

# wordList_english   = re.sub("[^\w]", " ",  english_train).split()
print("Words after Preprocessing",english_train)
total_words = len(english_train)
```
