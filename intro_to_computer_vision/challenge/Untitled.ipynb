```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
def convert_to_grayscale(im):
    grayImage = np.zeros(im.shape)
    R = im[:,:,0]*0.299
    G = im[:,:,1]*0.587
    B = im[:,:,2]*0.114
    avg = (R + G + B)
    plt.imshow(avg)
```
