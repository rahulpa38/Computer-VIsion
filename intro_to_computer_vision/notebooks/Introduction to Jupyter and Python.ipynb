## 6. Computery-Vision Stuff

This is a computer vision course, so we'll often use numpy and matplotlib to manipulate and visualize image arrays. Matplotlib includes a method `imread` that allows us to import image files into numpy arrays. Note that to import jpgs you may need to install the [Python Image Library (PIL)](https://pillow.readthedocs.io/en/stable/installation.html).

`````python
im = imread('data/sample_image.jpg')
`````

`````python
#Let's have a look at our image data.
im
`````

`````python
im.shape
`````

Our image is 256 by 256 pixels, and the 3 here corresponds to the red, green, and blue color channels of our image. 

`````python
im.dtype
`````

Our image array is storing our pixel values as unsigned 8 bit integers. We can display our image using matplotlib's `imshow` method.

`````python
fig = figure(0, (8,8))
imshow(im)
`````

We can also visualize the red, green, and blue color channels of our image independently:

`````python
fig = figure(0, (22,6)) #Allows us to control the size of figures
colors = ['red', 'green', 'blue']
for i in range(3):
    fig.add_subplot(1,3,i+1)
    imshow(im[:,:,i], vmin = 0, vmax = 255)
    title(colors[i])
    colorbar()
`````

- It's worth thinking through what the python code above is doing - we'll be using code patters like this *alot* in this course. 
- If you get stuck, it's best to go line by line - break apart the cell above into one line per cell, and investiage the inputs and outputs for each line. 
- We're visualizing each image color channel seperately. 
- Note that imshow does something different when we pass in a 2d vs 3d array:

`````python
im.shape
`````

`````python
imshow(im)
`````

`````python
im[:,:,0].shape
`````

`````python
imshow(im[:,:,0])
`````

Matplotlib plots our image using it's own [colormap](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html) - the default colormaps is the viridis colormap - we can add a colorbar that show which pixel values are mapped to which colors with the `colorbar()` method:

`````python
imshow(im[:,:,0])
colorbar()
`````

`````python
#There are also lots of other fun colormaps.
imshow(im[:,:,0], cmap = 'jet')
colorbar()
`````

Another really helpful thing to do is visualize a histogram of our image intensity values, like this:

`````python
fig = figure(0, (10,6))
hist(im.ravel(), 100);
grid(1)
`````

We can also do this independently for our 3 channels:

`````python
colors = ['r', 'g', 'b']
fig = figure(0, (18, 6))
for i, color in enumerate(colors):
    fig.add_subplot(1,3,i+1)
    hist(im[:,:,i].ravel(), 100, facecolor = color)
    grid(1); title(color)
`````

Based on our histograms, we can see that our image has more bright blues - this is likeley the sky. We can validate this by plotting a mask where the blue color channel exceeds some threshold.

`````python
threshold = 150

fig = figure(0, (18, 9))
fig.add_subplot(1,2,1)
imshow(im)

fig.add_subplot(1,2,2)
imshow(im[:,:,2] > threshold)
`````
